Morning Transactions

```python
morning_list = assigmnent_morning.groupby(['Transaction'])['Item'].apply(list).values.tolist()
encoder = mlxtend.preprocessing.TransactionEncoder().fit(morning_list)

encoded_data = encoder.transform(morning_list)

morning_trans = pd.DataFrame(encoded_data, columns = encoder.columns_)

print(morning_trans.head())
```

```python
morning_trans.shape
```

```python
morning_trans.sum().sort_values(ascending = False)[:10]
```

```python
# just for trying 
frequent_itemsets_morning = mlxtend.frequent_patterns.apriori(morning_trans, min_support = 0.001, max_len = 4, use_colnames = True)
frequent_itemsets_morning
```

```python
# Creating itemsets # Support values with max_len = 3

supp_test = [0.5, 0.3, 0.2, 0.4, 0.1, 0.05, 0.01, 0.005, 0.001,]


# 0.001 with  325 itemsets is good choice

for supp in supp_test:
    frequent_itemsets_morning = mlxtend.frequent_patterns.apriori(morning_trans, min_support = supp, max_len = 3, use_colnames = True)
    print('Number of frequent itemsets with support threshold', supp, 'is', frequent_itemsets_morning.shape[0])
```

```python
#with max length= 2 the maximum number of frequent itemeset is three 
```

```python
# Creating itemsets # Support values with max len = 4

supp_test = [0.5, 0.3, 0.2, 0.4, 0.1, 0.05, 0.01, 0.005, 0.001,]

# 0.001 with 334 itemsets is an okay choice

for supp in supp_test:
    frequent_itemsets_morning = mlxtend.frequent_patterns.apriori(morning_trans, min_support = supp, max_len = 4, use_colnames = True)
    print('Number of frequent itemsets with support threshold', supp, 'is', frequent_itemsets_morning.shape[0])
```

```python
#min_support = 0.001 with 334 itemsets and max_len = 4  is good choice
```

```python
frequent_itemsets_morning = mlxtend.frequent_patterns.apriori(morning_trans, min_support = 0.001, max_len = 4, use_colnames = True)
frequent_itemsets_morning
```

```python
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.1)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.2)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.3)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.4)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.5)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.6)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.7)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.8)))
print(len(mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.9)))
```

```python

rules_morning = mlxtend.frequent_patterns.association_rules(frequent_itemsets_morning, metric = "confidence", min_threshold = 0.3)
rules_morning 
```

```python
## the extracted rules for morning transaction contain useful information . Most frequent items such as coffee, bread and
#pastry were obeserved from the extracted rules. It is very ilkely that one could purchase coffee with muffin or pastry 
```

```python
## Recoomanandtion for Eggs purchase in the morning
selection_morning = rules_morning['antecedents'].apply(lambda m: 'Eggs' in m)
rules_morning[selection_morning]

## bread is recommeded for someone who bought eggs in the morning. It very likely that someone could buy bread after
#purchasing eggs to make breakfast
```
